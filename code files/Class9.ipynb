# Numpy:
Numpy is a low level library written in C & Fortrand for high level mathematical functions. Numpy is use to overcomes the problem of running slower algorithms on running python by using multidimensional arrays and functions that operate on arrays.Algorithm can expressed as a function on arrays, allowing the algorithms to be run quickly.



## Why NumPy is Faster than Python Lists:


###1. Continuous Memory (Compact Storage)

Python List → Stores elements in different places in memory and only keeps references.

NumPy Array → Stores elements next to each other in memory (like a C array).
This makes access much faster.

```python
#import numpy
import numpy as np
```

```python

py_list = [1, 2, 3, 4, 5]
np_array = np.array([1, 2, 3, 4, 5])

print("Python list element addresses:")
for x in py_list:
    print(id(x))

print("\nNumPy array element addresses:")
print(np_array.__array_interface__['data'])
```

###2. Vectorization (No Loops in Python)

Python List → You must use loops.

NumPy Array → Can apply operations to the whole array at once.
Internally, NumPy uses fast C code, not slow Python loops.

```python
# Python list with normal for loop
list_data = [1, 2, 3, 4, 5]
list_result = []
for x in list_data:
    list_result.append(x * 2)
print("List result:", list_result)

# NumPy array (vectorized)
np_data = np.array([1, 2, 3, 4, 5])
np_result = np_data * 2
print("NumPy result:", np_result)
```

###3. Same Data Type (Homogeneous Data):

Python List → Can store mixed types, e.g. [1, "hello", 3.5].
That means Python must check the type of each element again and again.
NumPy Array → Stores only one compact data type (common type of all elements). This makes operations much faster.

```python

py_list = [1, '2', 3.5, 4,True]
np_array = np.array([1, 2, 3, 4])


types_list = []
for x in py_list:
    types_list.append(type(x))

print("Python list types:", types_list)
print("NumPy array type:", np_array.dtype)
```

###What happens if we put mixed data types in a NumPy array?

Unlike Python lists (which can hold elements of different types without any issue),
NumPy arrays always enforce a single data type for all elements.If you mix different types, NumPy will automatically convert all elements to one common type that can fit everything.

```python
#Example 1 ---> int + str

arr = np.array([1, 2, "hello", 3])
print(arr)
print("Array dtype:", arr.dtype)


# The integers (1, 2, 3) were converted into strings ('1', '2', '3').
# NumPy decided the final type should be string, because a string can represent both text and numbers.
# <U11 means “Unicode string of maximum length 21”.
```

```python
# Example 2 ----> float + int
arr = np.array([1, 2, 3.5, 4])
print(arr)
print("Array:", arr)
print("Array dtype:", arr.dtype)

# The integers (1, 2, 4) were converted into floats (1.0, 2.0, 4.0).
# NumPy chose float64 because floats can hold decimal values, and they can also hold integers.
```

```python
# Example 3 -----> int + bool
arr = np.array([1, 0, True, False])
print(arr)
print("Array:", arr)
print("Array dtype:", arr.dtype)

# True was converted to 1, and False was converted to 0.
# NumPy chose int64, since integers can represent both numbers and booleans.
```

# List Vs Numpy-Array:

* Less Memory  

```python
import sys
```

```python
# Less Memory ------> Numpy takes less memory than python list
py_list = list(range(1000))
np_array = np.arange(1000)

print("Python list size:", sys.getsizeof(py_list), "bytes")
print("NumPy array size:", np_array.nbytes, "bytes")
```

* Faster

```python
import time
```

```python
#Faster -----> Numpy array operational speed much faster than list oerations
# Python list
py_list = list(range(1_000_000))
start = time.time()
py_result = []
for x in py_list:   # loop operation
    py_result.append(x * 2)
end = time.time()
print("Python list time:", end - start, "seconds")

# NumPy array
np_array = np.arange(1_000_000)
start = time.time()
np_result = np_array * 2   # Vectorized operation
end = time.time()
print("NumPy array time:", end - start, "seconds")
```

```python
# Multidimensional matrics
arr13=np.arange(24).reshape(6,4) # reshape : converts the array into given matrics size (roes,columns)
print(arr13)

#indexing
print(arr13[2])  # retrun row


#slicing
print(arr13[2:5]) #return rows

#slcing rows & columns ------> arr(slice rows , slice column)
print(arr13[:,2]) #return rows & column
print(arr13[:,1:4]) # retrun rows & columns
print(arr13[4:6,2:4]) # retrun rows & columns
```

```python
#multiblock & multidimensional matrics
arr14=np.arange(32).reshape(2,4,4)  #(block,row,column)
print(arr14)

#indexing
print(arr14[0])        # First block
print(arr14[1])        # Second block

print(arr14[0][0])     # first block first row
print(arr14[1][0])     # second block first row

print(arr14[0][0][1])  # frist block row 1 col 2
print(arr14[1][0][1])  # Second block row 1  col 2

#slicing
print(arr14[:][0][0])   # All blocks, first row, first col
print(arr14[:][1][1])   # All blocks, second row, second col

print(arr14[0][:][:])   # Entire first block
print(arr14[1][:][:])   # Entire second block

print(arr14[:][:][0])   # All blocks, all rows
print(arr14[:][:][1])   # All blocks, all rows
```

#Broadcasting

Broadcasting refers to the ability of numpy to treat arraysof different shapes during arthematic operations. Arthematic operations on arrays are usually done on corresponding elements, when two arrays are of exactly the same shape,then these operations are smoothly performed.

If the dimensions of two arrays are dissimilar, element-to-element operations are not possible. However, operations on arrays of on-similar shapes is still possible in Numpy, because of the broadcasting capability. The smaller array broadcast to the size of the larger array so that they have compatible shapes.

## Presentation of Broadcasting
numpy_broadcasting.png

```python
# Example
arr1=np.arange(9).reshape(3,3)
arr2=np.arange(3).reshape(1,3)

print("Array 1:\n",arr1)
print("Array 2: ",arr2)

print("Broad Casting:\n",arr1+arr2)
```

# Rules For Broadcasting

```python
# 1. if x=m & y=n, operation will execute
a1=np.arange(16).reshape(2,8)
a2=np.arange(8,24).reshape(2,8)

print("a1\n",a1)
print("a2\n",a2)
print("broadcoacting\n",a1+a2)
```

```python
# 2. if x=1 & y=n then also operation will execute (same correspondance)
a3=np.arange(4).reshape(1,4)
a4=np.arange(12).reshape(3,4)

print("a3\n",a3)
print("a4\n",a4)
print("broadcoacting\n",a3+a4)


a5=np.arange(4).reshape(4,1)
a6=np.arange(12).reshape(4,3)

print("a5\n",a5)
print("a6\n",a6)
print("broadcoacting\n",a5+a6)
```

```python
# 3. if x=1 & y!=n then operation will not execute
a7=np.arange(3).reshape(1,3)
a8=np.arange(16).reshape(4,4)

print("a7\n",a7)
print("a8\n",a8)
print("broadcasting\n",a7+a8)
```

```python
# 4. if (1, n) & (m, 1) then if will become (m,n), operation will execute
a9=np.arange(4).reshape(1,4)
a10=np.arange(4).reshape(4,1)

print("a9\n",a9)
print("a10\n",a10)
print("broadcasting\n",a9+a10)
```

```python
# 5. if x=1 & y=1, then operation will also execute
a11=np.arange(1).reshape(1,1)
a12=np.arange(16).reshape(4,4)

print("a11\n",a11)
print("a12\n",a12)
print("broadcasting\n",a11+a12)
```

```python
# 6. if they are of different dimensions
a13=np.arange(4)
a14=np.arange(24).reshape(6,4)

print("a13\n",a13)
print("a14\n",a14)
print("broadcasting\n",a13+a14)
```
