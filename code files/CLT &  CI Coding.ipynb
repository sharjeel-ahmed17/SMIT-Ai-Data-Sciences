# 1. Central Limit Theorem

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
#plotting distribution of fare column
df['fare'].plot(kind='kde')
```

```python
# sample size = 50 -->  100 times
samples = []
for i in range(120):
    samples.append(df['fare'].sample(50).tolist())
```

```python
samples
```

```python
len(samples)
```

```python
np.array(samples).shape
```

```python
samples=np.array(samples)
```

```python
samples
```

```python
# mean of each sample
sampling_means=samples.mean(axis=1)
```

```python
# ampling distribution of sample mean
sns.kdeplot(sampling_means)
```

```python
# calculating samling mean
sampling_means.mean()
```

```python
df['fare'].mean()
```

# Confidence Interval

```python
samples=[]
stds=[]
for i in range(10):
  x=df['fare'].sample(30)
  stds.append(x.std())
  samples.append(x.tolist())
```

```python
samples=np.array(samples)
```

```python
samples.shape
```

```python
# mean of each sample
sampling_means=samples.mean(axis=1)
```

```python
sampling_means
```

```python
sample_stds=np.mean(stds)
```

```python
sample_stds
```

```python
sns.kdeplot(sampling_means)
```

```python
#t test formula
lower_limit=sampling_means.mean()-2.04*(sample_stds/np.sqrt(30))
upper_limit=sampling_means.mean()+2.04*(sample_stds/np.sqrt(30))
```

```python
print("Range : ",(lower_limit,upper_limit))
```
