## Filtering Data:
It is als known as masking in which we ask a question to data frame and it returns answer. This concept is uefull for data analysis and decisions.

```python
# how many matches in hyderabad
df[df['city']=='Hyderabad'].shape[0]
```

```python
# how many matches in hyderabad after 2010 start
mask1=df['city']=='Hyderabad'
mask2=df['date']>'2010-01-01'
df[mask1 & mask2].shape[0]
```

## Drop Duplicates:
A very powerfull function use to drop duplicates row by row from the data.

```python
# for all data
df.drop_duplicates()
```

```python
# single column ----> drop dubicates on basis of city
df.drop_duplicates(subset=['city'])
```

```python
# multiplecolumns -----> drop duplicates on the basis of city and season occurances
df.drop_duplicates(subset=['city','season'])
```

```python
# who is the winner of each ipl season
df.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')
```

## Groupby Function:
The groupby function in Pandas is used to split data into groups based on one or more columns, then apply an operation to each group.

```python
#how many matches in each season
df.groupby('season')['date'].count()
```

```python
#toss impact
df.groupby('toss_decision')['winner'].count()
```

```python
# how many matches eahc team win in each city
df.groupby(['city','winner']).size().reset_index(name='matches_won')
```
