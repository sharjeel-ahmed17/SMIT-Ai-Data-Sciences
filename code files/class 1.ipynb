#Variables

**variables are container for future use**

```python
name='Abdul Rehman'

print(name)
```

1. No Variable declaration in python.
2. Python supports dynamic typing means no need of declare data type .
3. Static typing in which we have to declare the variable data type first
4. python supports Dynamic Binding feature means that Can change variable data type again and again, variable not bounded with one data type.
5. Static binding means that you can't change the data type of variable after declaration.

```python
# Dynamic Binding

name='Abdul Rehman'
print (name)

name=6
print(name)

name=True
print(name)
```

```python
# special Syntax ----> declare more vriables in one line

a=5;b=6;c=4

print(a)
print(b)
print(c)

a,b,c=7,8,9

print(a)
print(b)
print(c)

a=b=c=10
print(a)
print(b)
print(c)
```

#Identifiers

Python identifier is a name used to identify a variable, function, class, module, module or other object.

### Rules for setting identifier

* Can start with alphabate or _
* Followed by 0 or more letter, _ and digits
* Keywords cannot be use as identifier

```python
_="Name"
print(_)

__='Abdul'
print(__)
```
