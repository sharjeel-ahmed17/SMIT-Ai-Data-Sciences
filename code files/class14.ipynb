# Data Visualization Using Matplotlib

## Two Types of Data

- Numerical Data: Based on continuous numbers
- Categorical Data: Based on Groups

**Note:** If you lot the data on single column is called univarient analysis, when plot on two columns together is called bi-varient analysis. And when you plot graph on more thn two columns at a time it is called multivarient analysis.

##Types of analysis based on the number of variables:

*  **Univariate Analysis:** Involves one variable only,Used to understand the distribution, central tendency, and spread of a single variable.
* **Bivariate Analysis:** Involves two variables, Used to study the relationship or comparison between two variables.
* **Multivariate Analysis:** Involves three or more variables, Used to understand complex relationships among multiple factors.

## 2. Scatter Plots

A scatter plot is a type of 2D graph that shows the relationship between two numerical variables by displaying data points as dots on an X and Y axis. Each dot represents one observation, with its position determined by the values of the two variables. It helps identify patterns, trends, and correlations between the variables.

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case: Finding correlation

```python
# plt.scatter on data
df = pd.read_csv('/content/drive/MyDrive/python-Saylani/batter.csv')
df = df.head(50)
df
```

* Subplots

A subplot is a feature that allows multiple plots to be displayed within a single figure. It divides the figure into a grid of rows and columns, enabling you to visualize multiple graphs side by side for easy comparison of different datasets or relationships.

```python
#multiple plots using grid, side by side ---> plotting (2x1) grid

#fig ---> store figure where graph plot
#ax ----> save array for plot
fig, ax = plt.subplots(nrows=2,ncols=1,figsize=(10,6))

ax[0].scatter(df['avg'],df['strike_rate'],color='red')
ax[1].scatter(df['avg'],df['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')

plt.show()
```

```python
#plotting (2x2) grid ----> 2D Array
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(df['avg'],df['strike_rate'],color='red')
ax[0,1].scatter(df['avg'],df['runs'])
ax[1,0].hist(df['avg'])
ax[1,1].hist(df['runs'])

plt.show()
```

* 3D Scatter Plots:

A 3D scatter plot is a graphical representation that shows the relationship between three numerical variables in a three-dimensional space. Each point is plotted using its X, Y, and Z coordinates, allowing you to visualize patterns, trends, and correlations among three variables simultaneously.

```python
fig = plt.figure(figsize=(6,10))

ax = plt.subplot(projection='3d')

ax.scatter3D(df['runs'],df['avg'],df['strike_rate'])
ax.set_title('IPL batsman analysis')

ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('SR')

plt.show()
```

##3. Bar chart:
A bar chart is a graphical representation used to compare categorical and numerical data using rectangular bars. The categorical variable is shown on the X-axis, and the numerical variable on the Y-axis. The length or height of each bar represents the value or frequency of that category, making it easy to compare different groups visually.

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

* Simple Bar Chart

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

* horizontal bar chart

```python
plt.barh(colors,children,color='green')
```

```python
df = pd.read_csv('/content/drive/MyDrive/python-Saylani/batsman_season_record.csv')
df
```

```python
# barchart on batsman data ----> overlaping each other
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
#multiple bar plots of each category ---> side by side
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem ----> bigname of categories
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

* Stacked Bar Chart
image.png

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

A pie chart is a circular graph that shows how different categories contribute to a whole. Each slice represents the proportion or percentage of a numerical value for a categorical variable, making it ideal for (categorical + numerical) data.

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case : To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/drive/MyDrive/python-Saylani/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
