## Create Dataframe

```python
# create dataframe using list(2D)
student_data=[
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
# create dataframe using dictionary
student_dict={
    'Name':['Abdul Rehman','Abdullah','Nasir','Ali','Ameen','Umer'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

student=pd.DataFrame(student_dict)
student.set_index('Name',inplace=True)
```

# Select Rows From Dataframe

* **iloc** : Searches using index position
* **loc** : Searches using label position

### 1. iloc

```python
#single row using iloc
movies.iloc[5] #series
```

```python
# multiple rows using iloc
movies.iloc[0:5] #dataframe
```

```python
#multiple indexes
movies.iloc[[0,2,4]] #dataframe
```

### 2. loc

```python
#single row using loc
student.loc['Abdul Rehman']
```

```python
#multiple rows using loc
student.loc['Abdul Rehman':'Ali']
```

```python
#multiple indexes
student.loc[['Abdul Rehman','Ali']]
```

## Selecting Rows & columns using iloc & loc

```python
movies.iloc[0:3,0:3] #[rows,columns]
```

```python
movies.loc[0:3,'title_x':'title_y'] #[rows,columns]
```

## Filtering a dataframe (Practise)

```python
ipl.head(2)
```

```python
#find all the final winners
new_df=ipl[ipl['MatchNumber']=='Final']
new_df[['Season','WinningTeam']]

# code different same result
ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
# how many matches have occured super over
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# how many matches has csk wins in kolkata
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam']=="Chennai Super Kings")].shape[0]
```

```python
# how many time toss winner win the match in percentage
ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]/ipl.shape[0]
```

```python
movies.head(2)
```

```python
# total movies with higher rating than 8 and vote>10000
movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)].shape[0]
```

```python
# total action movies with rating higher than 7.5
mask1=movies['genres'].str.split('|').apply(lambda x: 'Action' in x)
mask2=movies['imdb_rating']>7.5
movies[mask1 & mask2].shape[0]
```
