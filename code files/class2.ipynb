# Type Conversion

You can convert from one type to another

### 2 Types of type conversion


*   Implicit: when python automatically perform type converion.
*   Explicit: when Programmer itself perform the type conversion.



```python
#implicit ----> python automatically convert the data type after addition

print(9+9.5) #int & float
print(5 + 6+7j)  # int & complex
print(2.5+9+5j) #  float & complex
```

```python
 # Explicit ----> Where python can't perform implicit type conversion

# float to int
print(int(7.8))

#int to float
print(float(9))

# string to int
print(int('45'))

#int to string
print((str(45)))

#float to strig
print(str(6.7))

#string to float
print(float('6.7'))

#int to bool
print(bool(0))

#bool to int
print(int(False))

#int to complex
print(complex(5))

```

**Note:** You cannot convert complex numbers into another number type.

```python
print(float(5+6j))
```

```python
# Not convertable

print(int('Karachi'))
print(float('Karachi'))
print(complex('Karachi'))
```

# Operators
Operators are use to perform oerations on variables and values.Python has following operators

*   Arthmatic Operators
*   Comparison Operators
*   Logical Opertors  
*   Assignment Operator
*   Identity Operators
*   Membership Operators
*   Bitwise Operators



```python
# Arthmatic Operators

x=7
y=2

#addition
print(x+y)

#subtration
print(x-y)

#multiplication
print(x*y)

#division
print(x/y)   #output in float by default

#percent
print(x%y)    #modulus

#floor division
print(x//y)      #integer division

#exponent
print(x**y)     #power
```

```python
# Comparison Operators
x=15
y=10

#greatr than , Less then
print(x>y)
print(x<y)

# greater than or equal to , less than or equal to
print(x>=y)
print(x<=y)

# Equal Equal to, Not Equal to
print(x==y)
print(x!=y)
```

```python
# Logicl Operators
x=True
y=False

# And operator
print(x and y)  # True, Flase = False

# Or operator
print(x or y) # True, False =True

# Not operator  =====> Shows negation
print(not x)


a=12
b=13
c=14

print(a<=b and c>=b)

print(a>=b or c<=b)

print(not(c>b and b>a))

print(not(b<c or a>c))
```

```python
# Assignment Operators

a=10
print(a)

a+=1
print(a)  #a=a+3

a-=1
print(a)

a*=2
print(a)

a**=2
print(a)

a/=2
print(a)

a%=2
print(a)

a//=2
print(a)

```

```python
# Identitiy operators -----> used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location. Output (True or False)

a=3
b=3

#is operator
print(a is b)

#is not operator
print(a is not b)


x="Hello"
y="World"

print(x is y)
print(x is not y)

x="Hello-world"
y="Hello-world"
print(x is y)
print(x is not y)

m=[1,2,3,4]
n=[1,2,3,4]
print(m is n)
print(m is not n)
```

```python
# Member ship Operators ----> shows the presence of a value in a sequence. Output (True or False)

x="Islamabad"

# in operator
print("I" in x)

#not-in operator
print("s" not in x)

x=[1,2,3,4,5]
print(3 in x)
print(6 not in x)

```

```python
# Bitwise Operators -----> used to compare (binary) numbers:

# & operator (and)
print(6 & 3)


"""
The & operator compares each bit and set it to 1 if both are 1, otherwise it is set to 0:

6 = 0000000000000110
3 = 0000000000000011
--------------------
2 = 0000000000000010
====================

"""
```

```python
# | operator  (or)
print(6 | 3)



"""
The | operator compares each bit and set it to 1 if one or both is 1, otherwise it is set to 0:

6 = 0000000000000110
3 = 0000000000000011
--------------------
7 = 0000000000000111
====================

"""
```
